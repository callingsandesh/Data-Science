## Linear regression

```python
slope,intercept=np.polyfit(raw_data['Healthy life expectancy'],raw_data['Logged GDP per capita'],1)

x=np.arange(45,80)
y=slope * x + intercept

plt.plot(x,y,color="red",alpha=0.8)
plt.scatter(raw_data['Healthy life expectancy'],raw_data['Logged GDP per capita'],)
plt.xlabel('Healthy life expectancy')
plt.ylabel('Logged GDP per capita')
```

The function np.polyfit() that is used here to get the regression parameters finds the optimal slope and intecept.It is optimizing the sum of the squares of the residuals(RSS).
